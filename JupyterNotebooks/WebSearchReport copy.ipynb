```python
import requests
from datetime import datetime, timedelta

def scrape_google_results(query, api_key, location, language, country, num_results, time_range, search_type):
    url = "https://serpapi.com/search.json"
    params = {
        "engine": "google",
        "q": query,
        "location_requested": location,
        "google_domain": f"google.{country}",
        "hl": language,
        "gl": country,
        "device": "desktop",
        "api_key": api_key,
        "num": 100,  # Maximum allowed by the API
        "start": 0,
        "tbs": time_range
    }

    # Add tbm parameter only if it's not an empty string
    if search_type:
        params["tbm"] = search_type

    all_results = []
    while len(all_results) < num_results:
        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()
            # The key for results varies based on the search type
            results_key = "news_results" if search_type == "nws" else "organic_results"
            results = data.get(results_key, [])
            if not results:
                break  # No more results available
            all_results.extend(results)
            params["start"] += len(results)  # Move to the next page
        else:
            print(f"Error: {response.status_code}")
            break

    return all_results[:num_results]  # Return only the requested number of results

def get_time_range(days):
    if days <= 0:
        return "qdr:h"  # Last hour (default for invalid input)
    elif days <= 1:
        return "qdr:d"  # Last 24 hours
    elif days <= 7:
        return "qdr:w"  # Last 7 days
    elif days <= 30:
        return "qdr:m"  # Last month
    elif days <= 90:
        return "qdr:m3" # Last 3 months
    else:
        return "qdr:y"  # Last year

# Your API key
api_key = "9d4a2ad1127f89d343c17a651bebd736c30d833a8be687058d4cbca8ddbdefc6"

# Search parameters
query = "Afvalverbrandingsinstallaties"
location = "Netherlands"
language = "nl"
country = "nl"
num_results = 150  # Specify the number of results you want
days_ago = 365  # Specify how many days back you want to search
search_type = ""  # Options: "" (empty string) for regular search, "nws" for news, "isch" for images, etc.

time_range = get_time_range(days_ago)
results = scrape_google_results(query, api_key, location, language, country, num_results, time_range, search_type)

if results:
    search_type_name = "news articles" if search_type == "nws" else "results"
    print(f"Retrieved {len(results)} {search_type_name} about '{query}' in {location} from the past {days_ago} days:")
    for i, result in enumerate(results, 1):
        print(f"\n{i}. {result.get('title', 'No title')}")
        
        if search_type == "nws":
            source = result.get('source', {})
            if isinstance(source, dict):
                print(f"   Source: {source.get('name', 'Unknown')}")
            else:
                print(f"   Source: {source}")
        
        print(f"   Date: {result.get('date', 'No date')}")
        print(f"   Link: {result.get('link', 'No link')}")
        
        if 'snippet' in result:
            print(f"   Snippet: {result['snippet']}")
else:
    print("No results found or an error occurred.")
```

```python
import requests
from bs4 import BeautifulSoup
import pprint
import time

def extract_article_content(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            paragraphs = soup.find_all('p')
            content = ' '.join([para.get_text() for para in paragraphs])
            return content
        else:
            return f"Error: Unable to fetch the article content. Status code: {response.status_code}"
    except Exception as e:
        return f"Error: {e}"

for index, result in enumerate(results):
    try:
        article_content = extract_article_content(result['link'])
        result['article_content'] = article_content
    except Exception as e:
        result['article_content'] = f"Error: {e}"
        print(f"Error: {e}")
    # Logging progress
    if index % 10 == 0:
        print(f"Processed {index+1}/{len(results)} items")

results
```

```python
def compile_and_count_characters(data):
    compiled_content = ""
    for entry in data:
        article_content = entry.get('article_content', '').replace('\n', ' ').replace('  ', ' ')
        entry['character_count'] = len(article_content)
        compiled_content += article_content + " "
    return compiled_content.strip(), data

compiled_content, updated_data = compile_and_count_characters(results)

filtered_results = [item for item in results if 1000 <= item.get('character_count', 0) <= 15000]
k = len(filtered_results)
print(k)
filtered_results
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma, FAISS
from langchain_openai import OpenAIEmbeddings
import faiss
import numpy as np
from langchain.docstore.document import Document

# Create a list of Document objects from your data
documents = [Document(page_content=item['article_content'], metadata={'index': i}) for i, item in enumerate(filtered_results)]

# Create embeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

# Create vector database
db = FAISS.from_documents(documents, embeddings)

# Define your query
query = "Wat kun je vertellen over de mening van het algemene publiek of van specifieke personen over afvalverbrandingsinstallaties in Nederland?"

# Perform similarity search
results = db.similarity_search_with_score(query, k=len(documents))

# Reorder your original list
reordered_results = [filtered_results[doc.metadata['index']] for doc, score in results]
```

```python
from Classes.Chatbot import ChatGPT

model_name = "gpt-3.5-turbo-16k"
temperature = 0
max_tokens = 900

summary_md = f"""
# {title_report_en}

"""
for result in reordered_results[:int(len(filtered_results)//3)]:

    chatbot = ChatGPT(
        model_name=model_name,
        temperature=temperature,
        max_tokens=max_tokens,
    )
    
    question =f"""
Your task is to extract and report specific stakeholders' opinion and position on the topic of {topic_of_interest} in {country}.
Please name the stakeholder, then, on the next line, provide a detailed and accurate summary of their opinion and position about {topic_of_interest}.
Here is the text from which you should extract this information: {result['article_content']}

Format your response as follows:
"Stakeholder: Name of person or organization.

Opinion  and  position:  Specific opinion of the stakeholder about {topic_of_interest} and their position on the topic."


Include in your output multiple stakeholders if multiple people express their opinion on the topic. Write your output in English.
    """

    response = chatbot.ask(question)

    title = result['title']
    link = result['link']
    single_output = f"""
## [{title}]({link})

{response}
"""

    summary_md += "\n" + single_output + "\n\n" + "---\n"

with open(f"KnowledgeBase/WebSearchReport/AVI/ReportEN.md", "w") as file:
    file.write(summary_md)
```
